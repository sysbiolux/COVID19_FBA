```python
df1 = pd.read_csv('data/time_series_study/SupplementaryData2_Calu3_polyA_series1_fpkm.tsv',sep='\t')
df2 = pd.read_csv('data/time_series_study/SupplementaryData3_Calu3_polyA_series2_fpkm.tsv',sep='\t')
df3 = pd.read_csv('data/time_series_study/SupplementaryData5_H1299_polyA_fpkm.tsv',sep='\t')
```

```python
dfs = [df1, df2, df3]
dfs = [df.set_index('gene_id||gene_name').drop(columns=['length']) for df in dfs]
df = dfs[0].join(dfs[1:])
```

```python
df['genes'] = df.index
df.index = df['genes'].apply(lambda x:x.split('||')[1])
df = df.drop(columns=['genes'])
```

```python
meta = pd.DataFrame()
meta['filename'] = df.columns
#remove SARS1 smaples
meta = meta.loc[~meta.filename.str.contains('S1'),:]
meta = meta.loc[~meta.filename.str.contains('untr'),:]
meta = meta.loc[~meta.filename.str.contains('detached'),:]

meta['Cell_Line'] = meta['filename'].apply(lambda x:x.split('_')[0])
#meta['phenotype'] = meta['filename'].apply(lambda x :'_'.join(x.split('_')[-1].split('-')[:-1]))

meta['Time_Point'] = meta['filename'].apply(lambda x:x[x.find('h-')-3 :x.find('h-')].split('-')[1])
#meta['Series'] = meta['filename'].apply(lambda x:x[x.find('polyA_'):x.find('polyA_')+15].split('-')[0].split('_')[1])
#meta['SARS_CoV'] = meta['filename'].apply(lambda x:x[x.find('_S'):x.find('_S')+1].split('_')[0])
#meta['detachedcells'] = meta['filename'].apply(lambda x:x.find('detachedcells'))
meta['Phenotype'] = meta.Cell_Line +'_' +meta.Time_Point.astype('str')+'h'
meta['Mock'] = meta['filename'].apply(lambda x:x.find('mock'))
meta.loc[meta.Mock>-1,'Mock']= 1
```

```python
meta.loc[meta['Mock']==1,'Mock']  = 'Mock' 
meta.loc[meta['Mock']==-1,'Mock']  = 'Infected' 
meta.head()
```
